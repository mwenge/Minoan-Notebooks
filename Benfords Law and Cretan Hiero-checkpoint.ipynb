```python
import json
from functools import reduce
import re

json_file = open('../LinearBInscriptions.js')
inscriptions = json.load(json_file)

prev_word = None
nums_for_entity = {}
for info in inscriptions:
    inscription = info[1]
    words = inscription["transliteratedWords"]
    for word in words:
        word = re.sub("(\[|\]|\n)", "", word)
        if not word[:1].isdigit():
            prev_word = word
            continue
        entity = prev_word
        if entity in nums_for_entity:
            nums_for_entity[entity] += [word]
        else:
            nums_for_entity[entity] = [word]
# flatten to a list of all numbers
all_nums = sum(nums_for_entity.values(), [])
# filter out anything that isn't an integer
all_nums = [n for n in all_nums if n.isnumeric()]
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter
%matplotlib inline

# get all first digits
first_digits = sorted([n[:1] for n in all_nums])
print([(i,first_digits.count(i)) for i in sorted(set(first_digits))])

# Plot the occurences of first digits
plt.hist(first_digits,bins=9)
plt.ylabel('Occurences')
plt.xlabel('First Digits');
```

```python
# Plot the occurences of first digits
plt.hist(first_digits, weights=np.ones(len(first_digits)) / len(first_digits),bins=9)
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.ylabel('Occurences')
plt.xlabel('First Digits');
```

```python
viable_commodities = {k:v for k,v in nums_for_entity.items() if len(v) > 100}
nrows = (int(len(viable_commodities)) / 4) + 1
ncols = 4

r=1
for c,nums in viable_commodities.items():
    # filter out anything that isn't an integer
    all_nums = [n for n in nums if n.isnumeric()]
    # get all first digits
    first_digits = sorted([n[:1] for n in all_nums])
    plt.subplot(nrows, ncols, r)
    # Plot the occurences of first digits
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.sans-serif'] = 'Aegean'
    plt.hist(first_digits,  bins=9)  # density=False would make counts
    plt.ylabel('Occurences')
    plt.xlabel(c + ' - First Digits');
    r+=1
plt.tight_layout()
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

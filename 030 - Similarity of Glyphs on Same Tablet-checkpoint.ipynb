```python
# Create a list of images and scribes from the individual letter images
# concatenated into 12 character strips

import os
import json
rootdir = 'letterimages'

imgs = []
glyph_dict = {}
for subdir, dirs, files in os.walk(rootdir):
    for f in files:
        letter, _, tablet = f[:-4].split('-')
        if tablet not in glyph_dict:
            glyph_dict[tablet] = {}
        if letter not in glyph_dict[tablet]:
            glyph_dict[tablet][letter] = [os.path.join(subdir,f)]
        else:
            c = glyph_dict[tablet][letter]
            glyph_dict[tablet][letter] = c + [os.path.join(subdir,f)]
```

```python
# Visualizing the data
multi_glyphs = {tablet + '-' + glyph: glyph_dict[tablet][glyph] 
                for tablet in glyph_dict for glyph in glyph_dict[tablet]
                if len(glyph_dict[tablet][glyph]) > 3
                }
```

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

def createStripFromLetters(letters):
    imgs = [ImageOps.grayscale(Image.open(l)) for l in letters]
    sizes = [i.size for i in imgs]
    strip_height = max(h for _,h in sizes)
    strip_width = sum(w for w,_ in sizes)
    
    strip = Image.new('RGB', (strip_width, strip_height))
    cw = 0
    for im,(w,h) in zip(imgs, sizes):
        strip.paste(im=im, box=(cw, 0))
        cw += w
        
    # Resize so height = 113 while keeping aspect ratio
    new_strip_height = 113
    height_fac = new_strip_height / strip_height
    new_width = int(strip_width * height_fac)
    size = new_width, new_strip_height
    return strip.resize((size), Image.ANTIALIAS)

i = 0
for t,imgs in multi_glyphs.items():
    img = createStripFromLetters(imgs)
    plt.figure()
    plt.imshow(img, cmap ='gray')
    i += 1
```

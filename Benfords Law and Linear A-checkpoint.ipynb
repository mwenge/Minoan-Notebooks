```python
import json

json_file = open('../../210-wordtags.js')
inscriptions = json.load(json_file)

nums_for_commodity = {}
for inscription in inscriptions:
    word_tags = inscription["tagsForWords"]

    commodity = ""
    for index, word_tag in enumerate(word_tags):
        word = word_tag["word"]
        tags = word_tag["tags"]
        word = word.replace('\U0001076b', '')
        if "logogram" in tags or "commodity" in tags:
            commodity = word
            continue
        if "number" not in tags:
            continue
        if "fraction" in tags:
            continue
        if commodity in nums_for_commodity:
            nums_for_commodity[commodity] += [word_tag["transliteratedWord"]]
        else:
            nums_for_commodity[commodity] = [word_tag["transliteratedWord"]]
# flatten to a list of all numbers
all_nums = sum(nums_for_commodity.values(), [])
# filter out anything that isn't an integer
all_nums = [n for n in all_nums if n.isnumeric()]
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter
%matplotlib inline

# get all first digits
first_digits = sorted([n[:1] for n in all_nums])
print([(i,first_digits.count(i)) for i in sorted(set(first_digits))])

# Plot the occurences of first digits
plt.hist(first_digits,bins=9)
plt.ylabel('Occurences')
plt.xlabel('First Digits');
```

```python
# Plot the occurences of first digits
plt.hist(first_digits, weights=np.ones(len(first_digits)) / len(first_digits),bins=9)
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.ylabel('Occurences')
plt.xlabel('First Digits');
```

```python
from scipy.stats import ks_2samp
import numpy as np

np.random.seed(12345678)
x = np.random.normal(0, 1, len(first_digits))
y = np.random.normal(0, 1, len(first_digits))
z = [int(i) for i in first_digits]

print(ks_2samp(x, y))
print(ks_2samp(x, z))
```

```python
viable_commodities = {k:v for k,v in nums_for_commodity.items() if len(v) > 40}
nrows = (int(len(viable_commodities)) / 4) + 1
ncols == 4

r=1
for c,nums in viable_commodities.items():
    # filter out anything that isn't an integer
    all_nums = [n for n in nums if n.isnumeric()]
    # get all first digits
    first_digits = sorted([n[:1] for n in all_nums])
    plt.subplot(nrows, ncols, r)
    # Plot the occurences of first digits
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.sans-serif'] = 'Aegean'
    plt.hist(first_digits,  bins=9)  # density=False would make counts
    plt.ylabel('Occurences')
    plt.xlabel(c + ' - First Digits');
    r+=1
plt.tight_layout()
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```
